## 3. Quick look at the data (2-3 real numbers)

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# 1. How many pages are actually declining?
decline_rate = (df["trend_direction"] == "down").mean()
print(f"Share of content declining: {decline_rate:.1%}")

# 2. Does position correlate with CTR the way you'd expect?
corr = df["avg_position"].corr(df["ctr"])
print(f"Correlation (avg_position vs ctr): {corr:.3f}")

# 3. Is there a real gap between high and low freshness in visibility?
by_freshness = df.groupby("freshness_tier")["impressions_90d"].median()
print(by_freshness)
```
