## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

My Lane as an ML Task is Classification.

It is classification because I have clear binary targets; declining vs not declining. A classifier is not my end goal, it's a tool to extract feature importance. I only care about which features the model leans on to make that prediction.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

The target for me would be the "trend-direction = 'down'" which tells whether a page is currently labeled as declining or not.

'trend_direction' is a derived bucket which has already been calculated by comparing current performance to a prior window and classified into "up/down/stable". I'm not measuring the raw decline myself, I'm inheriting a categorization someone already built.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

The metric for success for this lane would be ROC AUC which answers exactly my question: "if I had the model a declining page and a non-declining page, how often does it correctly tell them apart?", That's a direct measure of whether my chosen signal carry real separating information which is the whole point of the analysis.

What number means "good":

0.5 = model is guessing randomly, my signals explain nothing
0.6–0.7 = weak but real signal — worth reporting, with caveats
0.7–0.8 = solid signal — features clearly relate to decline
0.8+ = strong signal — I should be suspicious and check for leakage

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

The core reason is that decline isn't caused by one signal crossing one threshold, it is caused by combination of signals interacting.

A fixed rule can only check one or two conditions at a time.
